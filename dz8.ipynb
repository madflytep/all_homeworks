# Восстановление функции по зашумленным данным

Зададим первоначальную функцию, её первоначальные параметры и массив её значений и добавим к нему шум

```python
import numpy as np

def f(x, a, b, c):
    return c * np.exp(-((x - a) / b) ** 2)

x = np.arange(-2, 2, 0.25)
numbers = f(x, 0.5, 1, 2)

np.random.seed(123)
noise = 0.5 * np.random.normal(0, 1, x.size)
```

Отобразим настоящую функцию и зашумленные данные

```python
import matplotlib
import matplotlib.pyplot as plt
numbers = f(x, 0.5, 1, 2) + noise
plt.plot(x, f(x, 0.5, 1, 2))
plt.plot(x, numbers, 'o')
plt.show()
```

Попытаемся восстановить первоначальные параметры по обшему виду функции и данным с шумами

```python
from scipy.optimize import curve_fit
```

```python
popt, pooop = curve_fit(f, x, numbers, [1, 1, 1])
```

```python
plt.plot(x, f(x, *popt), '-o')
plt.plot(x, f(x, 0.5, 1, 2), '--o')
plt.show()
```

Сниним отображена полученная функция, оранжевым - первоначальная
